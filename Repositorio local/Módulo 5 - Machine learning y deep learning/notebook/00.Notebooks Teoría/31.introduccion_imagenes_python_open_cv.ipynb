### Leer imagenes con OpenCV

````python
img = cv2.imread(filename = "../Data/imagenes/imagen_prueba_2.jpeg")

type(img)

# OpenCV lee la imagen como un np.array
````

````python
img.shape
````

````python
plt.imshow(img)
plt.show()
````

La imagen fue leida correctamente por **`OpenCV`** como un **`np.array`**, pero los colores estan ordenados como **BGR**. Por defecto **`matplotlib`** espera una imagen en **RGB**, por lo que al momento de visualizarla veremos que los colores no son los esperados. Este comportamiento podemos cambiarlo transformando la imagen usando **`OpenCV`**.

````python
img_rgb = cv2.cvtColor(src = img, code = cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.show()
````

````python
# Si quisieramos abrir una imagen en escala de grises, podemos hacerlo usando OpenCV

img_grises = cv2.imread(filename = "../Data/imagenes/imagen_prueba_2.jpeg", flags = cv2.IMREAD_GRAYSCALE)

plt.imshow(img_grises, cmap = "gray")
plt.show()
````

### Cambiar el tamaño de imagenes con OpenCV

````python
img_rgb.shape
````

````python
# Usando un numero "fijo"

img_rgb_resize = cv2.resize(src = img_rgb, dsize = (1000, 400))

img_rgb_resize.shape
````

````python
plt.imshow(img_rgb_resize)
````

````python
# Usando un "ratio"

ancho_ratio = 0.1
alto_ratio = 0.2

img_rgb_resize = cv2.resize(src = img_rgb, dsize = (0, 0), fx = ancho_ratio, fy = alto_ratio)

plt.imshow(img_rgb_resize)
plt.show()
````

````python
img_rgb.shape
````

````python
img_rgb_resize.shape
````

### Voltear Imagenes

````python
# Vertical

v_img = cv2.flip(src = img_rgb, flipCode = 0)

plt.imshow(v_img)
plt.show()
````

````python
# Horizontal
v_img = cv2.flip(src = img_rgb, flipCode = 1)

plt.imshow(v_img)
plt.show()
````

````python
# Horizontal y Vertical
v_img = cv2.flip(src = img_rgb, flipCode = -1)

plt.imshow(v_img)
plt.show()
````

### Guardar imagenes con OpenCV

````python
cv2.imwrite("nueva_imagen_prueba2.jpg", v_img[:, :, ::-1])

# Esto guarda la imagen en "formato" de BGR y no en RGB
````
