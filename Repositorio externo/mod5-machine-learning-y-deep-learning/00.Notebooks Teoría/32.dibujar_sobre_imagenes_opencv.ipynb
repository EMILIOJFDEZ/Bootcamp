### Dibujar Rectangulos

**`cv2.rectangle()`**

- **img**: nombre de la variable de la imagen.
- **pt1**: ubicación (alto, ancho) de la esquina superior izquierda del rectangulo, debe ser una tupla.
- **pt2**: ubicación (alto, ancho) de la esquina inferior derecha del rectangulo, debe ser una tupla .
- **color**: color del rectangulo, debe ser una tupla de 3 elementos, es decir, un color escrito en RGB (numeros entre 0 y 255 para cada dimension).
- **thickness**: grosor del rectangulo. Si el valor es negativo la función llenará el rectangulo.
- **lineType**: Tipo de linea. _https://docs.opencv.org/4.x/d6/d6e/group__imgproc__draw.html_

```python
cv2.rectangle(img = img_vacia,
              pt1 = (384, 0),
              pt2 = (510, 128),
              color = (0, 255, 0),
              thickness = 3)

# Esta operación sobreescribe "img_vacia"

plt.imshow(img_vacia)
plt.show()
```

```python
cv2.rectangle(img = img_vacia,
              pt1 = (200, 200),
              pt2 = (300, 300),
              color = (0, 0, 255),
              thickness = -1)

plt.imshow(img_vacia)
plt.show()
```

### Circulos

**`cv2.circle()`**

- **img**: nombre de la variable de la imagen.
- **center**: centro del circulo (x, y), debe ser una tupla. 
- **radius**: radio del circulo.
- **color**: color del circulo, debe ser una tupla de 3 elementos, es decir, un color escrito en RGB (numeros entre 0 y 255 para cada dimension).
- **thickness**: grosor del circulo. Si el valor es negativo la función llenará el circulo.

```python
cv2.circle(img = img_vacia,
           center = (100, 100),
           radius = 50,
           color = (255, 0, 0),
           thickness = 5)

plt.imshow(img_vacia)
plt.show()
```

### Lineas

**`cv2.line()`**

- **img**: nombre de la variable de la imagen.
- **pt1**: ubicación (alto, ancho) del comienzo de la linea, debe ser una tupla.
- **pt2**: ubicación (alto, ancho) del final de la linea, debe ser una tupla .
- **color**: color de la linea, debe ser una tupla de 3 elementos, es decir, un color escrito en RGB (numeros entre 0 y 255 para cada dimension).
- **thickness**: grosor de la linea. En este caso el número debe ser positivo.

```python
cv2.line(img = img_vacia,
         pt1 = (100, 200),
         pt2 = (511, 511),
         color = (102, 255, 255),
         thickness = 5)

plt.imshow(img_vacia)
plt.show()
```

### Texto

**`cv2.putText()`**

- **img**: Nombre de la variable de la imagen.
- **text**: Texto.
- **org**: Donde comienza el texto (x, y), debe ser una tupla.
- **fontFace**: Tipo de fuente. _https://docs.opencv.org/4.x/d6/d6e/group__imgproc__draw.html_
- **fontScale**: Tamaño de la letra.
- **color**: Color del texto, debe ser una tupla de 3 elementos, es decir, un color escrito en RGB (numeros entre 0 y 255 para cada dimension). 
- **thickness**: Grosor del texto.
- **lineType**: Tipo de linea. _https://docs.opencv.org/4.x/d6/d6e/group__imgproc__draw.html_

```python
font = cv2.FONT_HERSHEY_SIMPLEX

cv2.putText(img = img_vacia, 
            text = "Hola Mundo",
            org = (10, 500),
            fontFace = font,
            fontScale = 2,
            color = (255, 255, 255),
            thickness = 2)
            #lineType = cv2.LINE_AA)

plt.imshow(img_vacia)
plt.show()
```

### Poligonos

Para dibujar un poligono necesitamos las ubicaciones de los vertices. Estos puntos deben estar en un **`np.array`** de tamaño (n, 1, 2) donde n es el número de vertices. El tipo de dato de estos números debe de ser int32.

**`cv2.polylines()`**

- **img**: Nombre de la variable de la imagen.
- **pts**: Puntos del poligono.
- **isClosed**: Booleano, cierra o no el poligono.
- **color**: Color del poligno.
- **thickness**: Grosor del poligono.

```python
img_vacia = np.zeros(shape = (512, 512, 3), dtype = np.int32)

vertices = np.array([[100, 300],
                     [200, 200],
                     [400, 300],
                     [200, 400]], np.int32)

pts = vertices.reshape((-1, 1, 2))

pts
```

```python
pts.shape
```

```python
cv2.polylines(img = img_vacia,
              pts = [pts],
              isClosed = True,
              color = (255, 0, 0),
              thickness = 5)

plt.imshow(img_vacia)
plt.show()
```

```python
img_vacia = np.zeros(shape = (512, 512, 3), dtype = np.int32)

cv2.polylines(img = img_vacia,
              pts = [pts],
              isClosed = False,
              color = (255, 0, 0),
              thickness = 5)

plt.imshow(img_vacia)
plt.show()
```
