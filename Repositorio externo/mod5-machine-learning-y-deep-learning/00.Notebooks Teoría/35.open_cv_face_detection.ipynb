## Detección de Rostros (Algoritmo Haar Cascades)

**`OpenCV`** cuenta con un algoritmo de detección de rostros llamado **`Haar Cascades`**.

Esta función puede reconocer rostros de imagenes y para poder utilizarlo hay que entrenarlo con los datos de: **`haarcascade_frontalface_default.xml`**.

En este ejemplo haremos reconocimiento de rostros, pero el algoritmo Haar Cascades puede reconocer otros tipos de objetos, solo hace falta entrenarlo con el **`.xml`** correspondiente:
_**https://github.com/opencv/opencv/tree/master/data/haarcascades**_

```python
import numpy as np

import matplotlib # Para ver la versión
import matplotlib.pyplot as plt

import cv2 
```

```python
# Versiones

print(f"numpy=={np.__version__}")
print(f"matplotlib=={matplotlib.__version__}")
print(f"opencv-python=={cv2.__version__}")
```

```python
# Imagenes

img_1 = cv2.imread(filename = "../Data/imagenes/foto1.jpg")
img_1 = cv2.cvtColor(src = img_1, code = cv2.COLOR_BGR2RGB)

img_2 = cv2.imread(filename = "../Data/imagenes/foto2.jpg")
img_2 = cv2.cvtColor(src = img_2, code = cv2.COLOR_BGR2RGB)

img_3 = cv2.imread(filename = "../Data/imagenes/foto3.jpg")
img_3 = cv2.cvtColor(src = img_3, code = cv2.COLOR_BGR2RGB)

img_4 = cv2.imread(filename = "../Data/imagenes/solvay_conference.jpg")
img_4 = cv2.cvtColor(src = img_4, code = cv2.COLOR_BGR2RGB)


plt.imshow(img_1)
plt.show()

plt.imshow(img_2)
plt.show()

plt.imshow(img_3)
plt.show()

plt.imshow(img_4)
plt.show()
```

```python
# Modelo

face_cascade = cv2.CascadeClassifier("../Data/haarcascade_frontalface_default.xml")
```

```python
# Función

def detect_face(img):
    
    face_img = img.copy()
  
    # .detectMultiScale() retorna las posiciones de las esquinas de un rectangulo
    face_rects = face_cascade.detectMultiScale(image = face_img, minNeighbors = 7)

    print(f"Se han detectado {len(face_rects)} caras")
    
    for x, y, w, h in face_rects: 
        
        cv2.rectangle(img = face_img,
                      pt1 = (x, y),
                      pt2 = (x + w, y + h),
                      color = (255, 0, 0),
                      thickness = 2) 
        
    return face_img
```

```python
img_1 = detect_face(img_1)

plt.imshow(img_1)
plt.show()
```

```python
img_2 = detect_face(img_2)

plt.imshow(img_2)
plt.show()
```

```python
img_3 = detect_face(img_3)

plt.imshow(img_3)
plt.show()
```

```python
img_4 = detect_face(img_4)

plt.imshow(img_4)
plt.show()
```
