### Vectores

Un vector es todo segmento de recta dirigido en el espacio.

<div align="center">
   
</div>

Cada vector posee las siguientes características que son:

- **Módulo:** Es la longitud o tamaño del vector. Para hallarla es preciso conocer el origen y el extremo del vector, para saber cuál es el módulo del vector, debemos medir desde su origen hasta su extremo.

    - El módulo de un vector $\overrightarrow{AB}$ es la longitud del segmento $AB$, se representa por $\left | \overrightarrow{AB} \right |$.



- **Dirección:** Viene dada por la orientación en el espacio de la recta que lo contiene.

- **Sentido:** Se indica mediante una punta de flecha situada en el extremo del vector, indicando hacia qué lado de la línea de acción se dirige el vector.

El sistema de referencia que se suele usar como norma general, es el Sistema de Coordenadas Cartesianas.

### Componentes de un vector

Un vector en un espacio (como en una hoja de papel) se puede descomponer en partes más pequeñas llamadas componentes.

Analogía: pensar en un vector como un desplazamiento que tienes que realizar caminando: puedes caminar una cierta distancia hacia el este (componente x) y otra hacia el norte (componente y).

Ejemplo en 2D:

Vector A: Imagina una flecha que va 3 unidades hacia la derecha (eje x) y 4 unidades hacia arriba (eje y).
Componentes de A: Aₓ = 3, Aᵧ = 4.


Cualquier vector puede ser considerado como resultado de la suma de vectores, cada uno de ellos en la dirección de uno de los ejes coordenados del sistema. 

$$\overrightarrow{AB} = x\overrightarrow{i}+y\overrightarrow{j}$$

Donde $x$ y $y$ son las magnitudes de los vectores auxiliares o componentes del vector $\overrightarrow{AB}$.

$\overrightarrow{i}$ y $\overrightarrow{j}$ son los vectores unitarios de esas componentes.

Este concepto se puede extender a **n-dimensiones**.

```python
plt.figure(figsize = (5, 5))

# Parametros: coordenadas base y largo en eje x y y 
# Descomponer el vector
plt.arrow(0, 0, 2.0, 2.0, head_width = 1, head_length = 1, color = "green") # Vector original
plt.arrow(0, 0, 2.0,   0, head_width = 1, head_length = 1, color = "black") # Vector unitario i
plt.arrow(0, 0,   0, 2.0, head_width = 1, head_length = 1, color = "brown") # Vector unitario j

plt.grid()
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.show()
```

```python
# 1. Definición de un Escalar
# Un escalar es simplemente un número. Por ejemplo:
escalar = 5
print(f"Escalar: {escalar}")

# 2. Definición de un Vector
# Un vector es una colección ordenada de escalares. Usaremos NumPy para crear vectores.
# Por ejemplo, un vector en 2D:
vector_2d = np.array([3, 4])
print(f"Vector 2D: {vector_2d}")

# Y un vector en 3D:
vector_3d = np.array([1, 2, 3])
print(f"Vector 3D: {vector_3d}")

# 3. Visualización de Vectores en 2D
# Utilizaremos Matplotlib para representar gráficamente el vector en 2D.

# Crear una figura y un eje
# sintaxis orientada a objetos en Matplotlib
fig, ax = plt.subplots()

# Dibujar el vector 2D como una flecha desde el origen (0,0) hasta (3,4)
ax.arrow(0, 0, vector_2d[0], vector_2d[1],
         head_width=0.2, head_length=0.3, fc='blue', ec='blue', length_includes_head=True)

# Configurar los límites de los ejes para una mejor visualización
ax.set_xlim(0, max(vector_2d[0], 5))
ax.set_ylim(0, max(vector_2d[1], 5))

# Añadir una cuadrícula
ax.grid(True)

# Añadir etiquetas y título
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title('Visualización de un Vector 2D')

# Mostrar el gráfico
plt.show()

# 4. Visualización de Vectores en 3D (Opcional)
# Para vectores en 3D, utilizaremos la funcionalidad 3D de Matplotlib.

from mpl_toolkits.mplot3d import Axes3D

# Crear una nueva figura para 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Dibujar el vector 3D como una flecha desde el origen (0,0,0) hasta (1,2,3)
ax.quiver(0, 0, 0, vector_3d[0], vector_3d[1], vector_3d[2],
          length=1.0, normalize=True, color='red')

# Configurar los límites de los ejes
ax.set_xlim([0, 4])
ax.set_ylim([0, 4])
ax.set_zlim([0, 4])

# Añadir etiquetas y título
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
ax.set_title('Visualización de un Vector 3D')

# Mostrar el gráfico
plt.show()
```

### Operaciones con vectores

#### Suma de vectores

Suma de Vectores: Para sumar dos vectores, simplemente sumas sus componentes correspondientes.

Ejemplo:

Vector A: (3, 4)

Vector B: (1, 2)

A + B: (3+1, 4+2) = (4, 6)

La suma de vectores es una operación matemática que combina dos o más vectores para obtener un nuevo vector. El vector resultado tiene lamisma dirección que la suma de las direcciones de los vectores originales.

```python
plt.figure(figsize = (5, 5))

# Vector negro más vector magenta

plt.arrow(0, 0, 2.0,   0, head_width = 0.3, head_length = 0.3, color = "black")
plt.arrow(2, 0,   2, 4.0, head_width = 0.3, head_length = 0.3, color = "magenta")

plt.arrow(0, 0,   4,   4, head_width = 0.3, head_length = 0.3, color = "green")

plt.grid()
plt.xlim(-1, 7)
plt.ylim(-1, 7)
plt.show()
```

#### Resta de vectores

Resta de Vectores: Similar a la suma, restas los componentes correspondientes.

Ejemplo:

Vector A: (3, 4)

Vector B: (1, 2)

A - B: (3-1, 4-2) = (2, 2)

```python
plt.figure(figsize = (6, 5))

# Vector negro menos vector magenta 

plt.arrow(0, 0, 2.0,    0, head_width = 0.5, head_length = 0.5, color = "black")
plt.arrow(0, 0,  -2, -4.0, head_width = 0.5, head_length = 0.5, color = "magenta")

plt.arrow(0, 0,   4,    4, head_width = 0.5, head_length = 0.5, color = "green")

plt.grid()
# acercar o alejar
plt.xlim(-7, 7)
plt.ylim(-7, 7)

plt.show()
```

#### Producto de un vector por un valor escalar

Multiplicación por un escalar: Multiplicar un vector por un número (escalar) cambia su magnitud pero no su dirección.

Ejemplo:

Vector A: (3, 4)

2 * A: (2 * 3, 2 * 4) = (6, 8)

El resultado de multiplicar un valor escalar $k$ por un vector $\overrightarrow{AB}$, expresado analíticamente por $k *\overrightarrow{AB}$, es otro vector con las siguientes características :

1. Tiene la misma dirección que $\overrightarrow{AB}$.
2. Su sentido coincide con el de $\overrightarrow{AB}$, si $k$ es un número positivo, y es el opuesto si $k$ es un número negativo.
3. El módulo es $k$ veces la longitud que representa el módulo de $\overrightarrow{AB}$ (Si $k$ es 0 el resultado es el vector nulo).

$$ k * \overrightarrow{AB}$$

$$ k = 3 $$

```python
plt.figure(figsize = (5, 5))

k = 3 

plt.arrow(0, 0,   2,   2, head_width = 0.3, head_length = 0.3, color = "red")

plt.arrow(0, 0, k*2, k*2, head_width = 0.3, head_length = 0.3, color = "black")

plt.grid()
plt.xlim(-7, 7)
plt.ylim(-7, 7)
plt.show()
```

```python
# Importar las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt

# 1. Definición de componentes de un vector

# Crear un vector en 2D
vector = np.array([4, 3])
print(f"Vector original: {vector}")

# 2. Acceso a componentes en NumPy

# Acceder a las componentes individuales
componente_x = vector[0]
componente_y = vector[1]
print(f"Componente X: {componente_x}")
print(f"Componente Y: {componente_y}")

# 3. Modificación de componentes

# Modificar la componente Y
vector[1] = 5
print(f"Vector después de modificar Y: {vector}")

# 4. Operaciones con componentes

# Crear otro vector para operaciones
otro_vector = np.array([1, 2])
print(f"Otro vector: {otro_vector}")

# Suma de vectores
suma = vector + otro_vector
print(f"Suma de vectores: {suma}")

# Resta de vectores
resta = vector - otro_vector
print(f"Resta de vectores: {resta}")

# Multiplicación por un escalar
escalar = 2
multiplicacion = vector * escalar
print(f"Vector multiplicado por {escalar}: {multiplicacion}")

# 5. Visualización de la descomposición de vectores

def plot_vector_with_components(vector, color='blue', label='Vector'):
    """
    Función para graficar un vector y sus componentes en 2D.

    Parameters:
    - vector: Arreglo de NumPy representando el vector [v_x, v_y].
    - color: Color de la flecha del vector.
    - label: Etiqueta del vector.
    """
    plt.figure()
    ax = plt.gca()

    # Dibujar el vector
    ax.arrow(0, 0, vector[0], vector[1],
             head_width=0.2, head_length=0.3, fc=color, ec=color, length_includes_head=True, label=label)

    # Dibujar las componentes
    # Componente X
    ax.plot([0, vector[0]], [0, 0], color='red', linestyle='--', label='Componente X')
    ax.arrow(vector[0], 0, 0, vector[1],
             head_width=0.2, head_length=0.3, fc='green', ec='green', length_includes_head=True, label='Componente Y')

    # Configurar los límites de los ejes
    max_val = max(abs(vector[0]), abs(vector[1])) + 2
    ax.set_xlim(-1, max_val)
    ax.set_ylim(-1, max_val)

    # Añadir líneas de los ejes
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)

    # Añadir cuadrícula
    ax.grid(True, which='both')

    # Añadir leyenda
    plt.legend()

    # Añadir etiquetas y título
    plt.xlabel('Eje X')
    plt.ylabel('Eje Y')
    plt.title('Descomposición de un Vector en Componentes')

    # Mostrar el gráfico
    plt.show()

# Graficar el vector original con sus componentes
plot_vector_with_components(vector, color='blue', label='Vector A')

# Graficar el vector modificado con sus componentes
plot_vector_with_components(otro_vector, color='purple', label='Vector B')

# Graficar la suma de los vectores con sus componentes
plot_vector_with_components(suma, color='orange', label='Suma de A + B')

# Graficar la resta de los vectores con sus componentes
plot_vector_with_components(resta, color='brown', label='Resta de A - B')

# Graficar el vector escalado con sus componentes
plot_vector_with_components(multiplicacion, color='cyan', label=f'Vector A x {escalar}')
```

### Producto escalar de dos vectores (Producto Escalar o Producto Punto)

El producto escalar es una operación que toma dos vectores y devuelve un número. Es útil para determinar cosas como el ángulo entre dos vectores.

Fórmula en 2D: A · B = Aₓ * Bₓ + Aᵧ * Bᵧ

Ejemplo:

A: (3, 4)

B: (1, 2)

A · B  --> 3 * 1 + 4 * 2 = 3 + 8 = 11

Se obtiene de la suma de los productos formados por las componentes de uno y otro vector. Es decir, dados dos vectores $\overrightarrow{AB}$ y $\overrightarrow{CD}$, expresados en un mismo sistema de coordenadas:
$$\overrightarrow{AB} = x_{\overrightarrow{AB}}*\overrightarrow{i}+y_{\overrightarrow{AB}}*\overrightarrow{j}$$
$$\overrightarrow{CD} = x_{\overrightarrow{CD}}*\overrightarrow{i}+y_{\overrightarrow{CD}}*\overrightarrow{j}$$


$$\overrightarrow{AB}\cdot \overrightarrow{CD}= x_{\overrightarrow{AB}}* x_{\overrightarrow{CD}}+y_{\overrightarrow{AB}}*y_{\overrightarrow{CD}}$$


El producto escalar también viene dado por:

$$\overrightarrow{AB}\cdot \overrightarrow{CD} = \left | \overrightarrow{AB} \right | \cdot \left | \overrightarrow{CD} \right | \cdot cos (\alpha)$$

Donde $\left | \overrightarrow{AB} \right |$ y $\left | \overrightarrow{CD} \right |$ son los módulos de sus vectores y $\alpha$ es el ángulo que se forma al unir ambos vectores.

Para calcular el producto escalar usaremos la función **np.dot()**.

```python
plt.figure(figsize = (5, 5))

plt.arrow(0, 0, 2, 4, head_width = 0.3, head_length = 0.3, color = "green")
plt.arrow(0, 0, 2, 2, head_width = 0.3, head_length = 0.3, color = "blue")

plt.grid()
plt.xlim(-1, 4)
plt.ylim(-1, 5)
plt.show()
```

```python
# Producto escalar 

AB = np.array([2, 4])
CD = np.array([2, 2]) 

print(np.dot(AB, CD))
print(AB @ CD)
```

```python
# Producto escalar 

AB = np.array([2, 3, 4])
CD = np.array([1, 4, 3])

np.dot(AB, CD)
```

Es posible calcular el ángulo entre dos vectores, ya que el producto escalar se puede hallar en función de sus módulos y del coseno del ángulo que forman mediante la siguiente fórmula:

$$\alpha =  arcos (\frac{\overrightarrow{AB}\cdot \overrightarrow{CD}}{\left | \overrightarrow{AB} \right | \cdot \left | \overrightarrow{CD} \right |}) $$

```python
producto = np.dot(AB, CD)
print(f"Producto punto: {producto}")

# modulo indica el tamaño o longitud de un vector, indica cuán grande es un vector
modAB = np.linalg.norm(AB)
modCD = np.linalg.norm(CD)
print(f"Modulo de AB: {modAB}")
print(f"Modulo de CD: {modCD}")

alpha = np.arccos(producto/(modAB*modCD))
print(f"Alpha: {alpha}")

degrees = np.degrees(alpha)
print(f"Degrees: {degrees}")
```

### Producto vectorial entre dos vectores (Producto Vectorial o Producto Cruz)

El producto vectorial es una operación que toma dos vectores y devuelve otro vector que es perpendicular a ambos. Sin embargo, en 2D, este concepto se simplifica y se puede considerar como un número que indica el "sentido de giro" entre dos vectores.

Nota: El producto vectorial es más común en 3D


El producto vectorial entre dos vectores, se define como un vector, donde su dirección es perpendicular al plano de los vectores, en el sentido del movimiento de un tornillo que gira hacia la derecha por el camino más corto del vector $\overrightarrow{AB}$ a $\overrightarrow{CD}$

$$
\left | \overrightarrow{AB} \text{ x } \overrightarrow{CD} \right | = \left | \overrightarrow{AB} \right | \cdot \left | \overrightarrow{CD} \right | \cdot sin (\alpha)
$$

Es posible calcular el ángulo entre dos vectores, ya que el producto vectorial se puede hallar en función de sus módulos y del seno del ángulo que forman mediante la siguiente fórmula:

$$
\alpha =  arcsin(\frac{\left | \overrightarrow{AB} \text{ x } \overrightarrow{CD} \right |}{\left | \overrightarrow{AB} \right | \cdot \left | \overrightarrow{CD} \right |})
$$

<div align="center">
   
</div>

**Producto vectorial de dos vectores de 2 dimensiones:**

$$\large{
\color{Green}{\overrightarrow{a} = 10\hat{i} - 3\hat{j}}\\
\color{Red}{\overrightarrow{b} = -2\hat{i} + 4\hat{j}}\\
\color{Green}{\overrightarrow{a}}\text{x}\color{Red}{\overrightarrow{b}}\color{Black} = ((\color{Red}{10})(\color{Green}{4}) - (\color{Green}{-3})(\color{Red}{-2}))\hat{k}\\
\color{Green}{\overrightarrow{a}}\text{x}\color{Red}{\overrightarrow{b}}\color{Black} = (40 - 6)\hat{k}\\
\color{Green}{\overrightarrow{a}}\text{x}\color{Red}{\overrightarrow{b}}\color{Black} = 34\hat{k}}
$$

Para calcular el producto vectorial o cruz usaremos **np.cross()**.

```python
# Producto vectorial 3-dim

AB = np.array([2, 3, 4])
CD = np.array([1, 4, 3])

producto_cruz = np.cross(AB, CD)
producto_cruz
```

```python
# Modulo producto cruz
modulo_prod = np.linalg.norm(producto_cruz)
print(f"Modulo producto cruz: {modulo_prod}")

# Modulo AB y CD
modAB = np.linalg.norm(AB)
modCD = np.linalg.norm(CD)
print(f"Modulo de AB: {modAB}")
print(f"Modulo de CD: {modCD}")

# Angulo
alpha = np.arcsin(modulo_prod / (modAB * modCD))
print(f"Alpha: {alpha}")

# Degrees
degrees = np.degrees(alpha)
print(f"Degrees: {degrees}")
```

```python
# Producto escalar 2-dim

AB = np.array([2, 4])
CD = np.array([2, 2]) 

producto_cruz = np.cross(AB, CD)
producto_cruz
```

```python
# Modulo producto cruz
modulo_prod = np.linalg.norm(producto_cruz)
print(f"Modulo producto cruz: {modulo_prod}")

# Modulo AB y CD
modAB = np.linalg.norm(AB)
modCD = np.linalg.norm(CD)
print(f"Modulo de AB: {modAB}")
print(f"Modulo de CD: {modCD}")

# Angulo
alpha = np.arcsin(modulo_prod / (modAB * modCD))
print(f"Alpha: {alpha}")

# Degrees
degrees = np.degrees(alpha)
print(f"Degrees: {degrees}")
```
