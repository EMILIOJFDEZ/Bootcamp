### Probabilidad Total

El Teorema de la probabilidad total nos permite calcular la probabilidad de un suceso a partir de probabilidades condicionadas.

**Ejemplo:**

Se dispone de tres cajas con bombillas. La primera contiene 10 bombillas, de las cuales hay cuatro fundidas; en la segunda hay seis bombillas, estando una de ellas fundida, y la tercera caja hay tres bombillas fundidas de un total de ocho. ¿Cuál es la probabilidad de que al tomar una bombilla al azar de cualquiera de las cajas, esté fundida?

<div align="center">
   
</div>

$$
\large{P(fundida) = \left(\frac{4}{10} \cdot \frac{1}{3}\right) + \left(\frac{1}{6} \cdot \frac{1}{3}\right) + \left(\frac{3}{8} \cdot \frac{1}{3}\right) =  0.31}
$$

## Teorema de Bayes

Desarrollado por el monje Thomas Bayes en el siglo XVII. El teorema de Bayes se utiliza para revisar probabilidades previamente calculadas cuando se posee nueva información.

$$
\Large {P(A_{i}|B) = \frac{P(A_{i}) \cdot P(B|A_{i})}{P(B)}}
$$

- $P(A_{i}|B)$: Probabilidad a posteriori de $A_{i}$ dado $B$ (lo que queremos calcular).
- $P(A_{i})$: Probabilidad a priori de $A_{i}$ (nuestra creencia inicial).
- $P(B|A_{i})$: Probabilidad condicional de $B$ dado $A_{i}$.
- $P(B)$: Probabilidad marginal de $B$.

**Ejemplo**:

Imagina que tienes dos cajas:

- **Caja A**: Contiene 10 bolas rojas y 10 bolas azules. (Total = 20)
- **Caja B**: Contiene 20 bolas rojas y 80 bolas azules. (Total = 100)

Sacas una bola al azar de una de las cajas sin mirar, y la observas. La bola es roja. Ahora, te preguntas: ¿de qué caja es más probable que haya salido la bola roja?

Intuitivamente, es más probable que la bola roja haya salido de la caja B, ya que tiene más bolas rojas que la caja A.

El **teorema de Bayes** nos permite calcular esta probabilidad de forma precisa.

Vamos a definir las variables:

- $A$: La bola roja salió de la caja A.
- $B$: La bola roja salió de la caja B.
- $E$: La bola sacada es roja. (Evento)

Tendremos que calcular el **teorema de Bayes** para ambas cajas:

$$\Large{P(A|E) = \frac{P(A) \cdot P(E|A)}{P(E)} \text{ y } P(B|E) = \frac{P(B) \cdot P(E|B)}{P(E)}}$$

Calculamos:

- $P(A)$ = Probabilidad a priori de que la bola haya salido de la caja A. Es $\frac{1}{2}$, ya que hay dos cajas.
- $P(B)$ = Probabilidad a priori de que la bola haya salido de la caja B. Es $\frac{1}{2}$, ya que hay dos cajas.
- $P(E|A)$: Probabilidad de que la bola sea roja si salió de la caja A. Es $\frac{10}{20}$, ya que hay 10 bolas rojas en la caja A.
- $P(E|B)$: Probabilidad de que la bola sea roja si salió de la caja B. Es $\frac{20}{100}$, ya que hay 20 bolas rojas en la caja B.
- $P(E)$: Probabilidad de que la bola sacada sea roja, sin importar de que caja haya salido. Es $P(A) \cdot P(E|A) + P(B) \cdot P(E|B) = \frac{1}{2} \cdot \frac{10}{20} + \frac{1}{2} \cdot \frac{20}{100}$

Sustituimos y resolvemos:

Para la caja A:

$$
\Large{P(A|E) = \frac{P(A) \cdot P(E|A)}{P(E)} = \frac{\frac{1}{2} \cdot \frac{10}{20}}{\frac{1}{2} \cdot \frac{10}{20} + \frac{1}{2} \cdot \frac{20}{100}} = \frac{2}{3}}
$$

Para la caja B:

$$
\Large{P(B|E) = \frac{P(B) \cdot P(E|B)}{P(E)} = \frac{\frac{1}{2} \cdot \frac{20}{100}}{\frac{1}{2} \cdot \frac{10}{20} + \frac{1}{2} \cdot \frac{20}{100}} = \frac{1}{3}}
$$

Según el **teorema de Bayes**, es más probable que salga de la caja A, al contrario de lo que habíamos intuido inicialmente.
