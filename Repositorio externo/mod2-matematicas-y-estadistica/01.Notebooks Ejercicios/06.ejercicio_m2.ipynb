```python
price = np.genfromtxt("../Data/idealista_madrid.csv", usecols=(0), delimiter=",", skip_header=1, encoding="utf-8", dtype=np.int64)
```

```python
baths = np.genfromtxt("../Data/idealista_madrid.csv", usecols=(1), delimiter=",", skip_header=1, encoding="utf-8", dtype=np.int64)
```

```python
neighborhood = np.genfromtxt("../Data/idealista_madrid.csv", usecols=(4), delimiter=",", skip_header=1, encoding="utf-8", dtype=np.str_)
neighborhood
```

```python
p80 = np.percentile(price, 80)
p80
```

```python
p20 = np.percentile(price, 20)
p20
```

```python
# Filtrar los barrios 20% más caros
barrios_mas_caros = neighborhood[price >= p80]
precios_mas_caros = price[price >= p80]
print(f"barrios_mas_caros: {barrios_mas_caros}")
print(f"precios_mas_caros: {precios_mas_caros}")
```

```python
barrios_mas_baratos = neighborhood[price <= p20]
precios_mas_baratos = price[price <= p20]
print(f"barrios_mas_caros: {barrios_mas_baratos}")
print(f"precios_mas_caros: {precios_mas_baratos}")
```

```python
from collections import Counter


def calcular_moda(datos):
    conteo = Counter(datos)
    moda = conteo.most_common(1)[0]  # Retorna el más común y su frecuencia
    return moda

moda_barrio, frecuencia_barrio = calcular_moda(neighborhood)

moda_mas_caros, frecuencia_mas_caros = calcular_moda(barrios_mas_caros)

moda_mas_baratos, frecuencia_mas_baratos = calcular_moda(barrios_mas_baratos)

# Imprimir resultados
print("Precio percentil 80 (más caro):", p80)
print("Precio percentil 20 (más barato):", p20)
print("\n20% más caros:")
for b, p in zip(barrios_mas_caros, precios_mas_caros):
    print(f"Barrio: {b}, Precio: {p}")

print("\n20% más baratos:")
for b, p in zip(barrios_mas_baratos, precios_mas_baratos):
    print(f"Barrio: {b}, Precio: {p}")

print("\nModa de todos los barrios:", moda_barrio, "(Frecuencia:", frecuencia_barrio, ")")
print("Moda de los barrios más caros:", moda_mas_caros, "(Frecuencia:", frecuencia_mas_caros, ")")
print("Moda de los barrios más baratos:", moda_mas_baratos, "(Frecuencia:", frecuencia_mas_baratos, ")")
```

```python


# Calcular moda usando np.unique
def calcular_moda(datos):
    valores, conteos = np.unique(datos, return_counts=True)
    indice_max = np.argmax(conteos)
    return valores[indice_max], conteos[indice_max]

# Moda de todos los barrios
moda_barrio, frecuencia_barrio = calcular_moda(neighborhood)

# Moda de los barrios más caros
moda_mas_caros, frecuencia_mas_caros = calcular_moda(barrios_mas_caros)

# Moda de los barrios más baratos
moda_mas_baratos, frecuencia_mas_baratos = calcular_moda(barrios_mas_baratos)

# Imprimir resultados
print("Precio percentil 80 (más caro):", p80)
print("Precio percentil 20 (más barato):", p20)
print("\n20% más caros:")
for b, p in zip(barrios_mas_caros, precios_mas_caros):
    print(f"Barrio: {b}, Precio: {p}")

print("\n20% más baratos:")
for b, p in zip(barrios_mas_baratos, precios_mas_baratos):
    print(f"Barrio: {b}, Precio: {p}")

print("\nModa de todos los barrios:", moda_barrio, "(Frecuencia:", frecuencia_barrio, ")")
print("Moda de los barrios más caros:", moda_mas_caros, "(Frecuencia:", frecuencia_mas_caros, ")")
print("Moda de los barrios más baratos:", moda_mas_baratos, "(Frecuencia:", frecuencia_mas_baratos, ")")
```

```python
from scipy import stats
# H0: Las casas en Recoletos cuestan lo mismo que en Hortaleza (media igual)
# H1: Las casas en Recoletos son más caras que en Hortaleza

barrio_x = "Recoletos"
barrio_y = "Hortaleza"
precios_x = price[neighborhood == barrio_x]
precios_y = price[neighborhood == barrio_y]

stat1, pvalue1 = stats.ttest_ind(precios_x, precios_y, alternative='greater')

print("\nContraste de hipótesis 1: Precios en", barrio_x, "vs", barrio_y)
print("Estadístico:", stat1, "P-valor:", pvalue1)
if pvalue1 < 0.05:
    print("Rechazamos H0: Las casas en", barrio_x, "son más caras que en", barrio_y)
else:
    print("No podemos rechazar H0: No hay evidencia de que las casas en", barrio_x, "sean más caras que en", barrio_y)
```

```python
# Contraste de hipótesis 2: Casas con >= 3 baños vs 1-2 baños

# H0: Las casas con >= 3 baños cuestan lo mismo que las de 1-2 baños (media igual)
# H1: Las casas con >= 3 baños son más caras que las de 1-2 baños
precios_3_banos = price[baths >= 3]
precios_1_2_banos = price[(baths == 1) | (baths == 2)]

stat2, pvalue2 = stats.ttest_ind(precios_3_banos, precios_1_2_banos, alternative='greater')

print("\nContraste de hipótesis 2: Casas con >= 3 baños vs 1-2 baños")
print("Estadístico:", stat2, "P-valor:", pvalue2)
if pvalue2 < 0.05:
    print("Rechazamos H0: Las casas con >= 3 baños son más caras que las de 1-2 baños")
else:
    print("No podemos rechazar H0: No hay evidencia de que las casas con >= 3 baños sean más caras que las de 1-2 baños")

```
