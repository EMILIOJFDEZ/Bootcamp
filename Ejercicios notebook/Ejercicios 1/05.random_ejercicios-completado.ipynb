## Ejercicios - random

```python
import random
```

### Ejercicio 1
- Crea una lista de 10 numeros enteros aleatorios entre el rango de 10 y 20 utilizando la función random.randint()

```python
''' Usamos una lista por comprensión para generar 10 números aleatorios entre 10 y 20,
random.randint(a, b) devuelve un número entero aleatorio en el rango [a, b] '''

lista_aleatoria = [random.randint(10, 20) for _ in range(10)]

# Mostramos la lista generada
print("Lista de números aleatorios entre 10 y 20 con random.randint():", lista_aleatoria)
```

### Ejercicio 2
- Crea una lista de 10 numeros enteros aleatorios entre el rango de 10 y 20 sin utilizar la función random.randint()

```python
# Creamos una lista vacía para guardar los números aleatorios
lista_aleatoria_sin_randint = []

""" Usamos random.uniform() para generar números de punto flotante aleatorios,
multiplicamos por 10 y convertimos a entero para obtener un rango entre 10 y 20 """
for _ in range(10):
    # random.uniform(10, 20) genera un número de punto flotante en el rango [10, 20)
    num = int(random.uniform(10, 20))  # Convertimos el número flotante a entero
    lista_aleatoria_sin_randint.append(num)

# Mostramos la lista generada
print("Lista de números aleatorios entre 10 y 20 sin random.randint():", lista_aleatoria_sin_randint)
```

### Ejercicio 3
- Genera una lista de 100 numero enteros aleatorios entre 0 y 100.
- Genera otra lista con solo los números pares de la primera lista.

- Extra: Hacer este ejercicio usando 1 solo bucle.

```python
# Generamos una lista de 100 números aleatorios entre 0 y 100
numeros_aleatorios = [random.randint(0, 100) for _ in range(100)]

# Usamos un solo bucle para generar la lista de números pares
# Si el número es par, lo agregamos a la lista de pares
numeros_pares = [num for num in numeros_aleatorios if num % 2 == 0]

# Mostramos ambas listas
print("Lista de 100 números aleatorios entre 0 y 100:", numeros_aleatorios)
print("Lista de números pares:", numeros_pares)
```
