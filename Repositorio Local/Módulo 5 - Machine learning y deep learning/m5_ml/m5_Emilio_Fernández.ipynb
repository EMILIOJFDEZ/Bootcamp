# Módulo 5: Machine learning y deep learning - Ejercicio de evaluación

## 1. Entendimiento del negocio y de los datos:
- Se analiza el dataset para entender qué información contiene.
- Se identifican las columnas relevantes y se eliminan las que no aportan valor (id, host_id, host_name).
- Se analiza la columna last_review y se decide cómo tratar los valores nulos.

## 2. Exploración y análisis de datos (EDA):
- Visualización de los datos con histogramas, boxplots, scatterplots y heatmaps de correlación.
- Identificación de patrones y relaciones entre variables.

## 3. Preprocesamiento de los datos:
- Datos numéricos: Imputación de valores nulos, escalado y transformación.
- Datos categóricos: Imputación de valores nulos, codificación con OrdinalEncoder o OneHotEncoder.
- Se prefiere usar Scikit-Learn en lugar de pandas para estas transformaciones.
- Posible reducción de dimensionalidad con PCA o SelectKBest.

## 4. Clustering:
- Se agrupan los datos en clusters utilizando K-Means o algún otro algoritmo de clustering.
- Se usa el cluster como nueva característica y se analiza su impacto.

## 5. Feature Selection:
- Se aplican métodos como SelectKBest o PCA para seleccionar las mejores características.
- Se comparan los resultados usando ambos métodos.

## 6. Modelado:
- Regresión: Se entrena un modelo para predecir price.
- Clasificación: Se entrena un modelo para predecir room_type.
- Se pueden probar distintos modelos de Scikit-Learn (Regresión Lineal, Random Forest, etc.).
- Opción de probar redes neuronales con TensorFlow-Keras.

## 7. Evaluación y comparación de modelos:
- Se validan los modelos con validación cruzada.
- Se comparan métricas como MAE, MSE para regresión y precisión, recall, F1-score para clasificación.
- Se pueden visualizar los resultados con gráficos como boxplots.

## 8. Despliegue (Opcional):
- Se empaqueta el modelo para producción.
- Se podría implementar con Flask, FastAPI o similar.

## Puntos adicionales opcionales:
- Transformaciones avanzadas de datos, como vectorizar name con TruncatedSVD.
- Uso de pipelines para automatizar preprocesamientos.
- SMOTE para manejar desbalanceo en room_type.
- Optimización de hiperparámetros con GridSearchCV.
